# Template Matching
***
Template matching is the simplest form of object detection. Template matching looks for exact copies of a template image by sliding the template over the image via 2D convolution, and will judge matches based on the similarity between the image neighborhood and the template image. The result is a grayscale image where the intesity value of each pixel shows how similar the template was to that pixel's neighborhood. The major issue with template matching is that it looks for exact copies, and does not respond well to changes in perspective. So if I wanted to use template matching for face detection, this method will be largely ineffective because changes in perspective, such as the turning of the person's face, will throw the algorithm off.

### Example 1 - Single Match
In this example, the goal is to find George Washington's face on the US quarter given the quarters image and the template image of his face. The matchTemplate() method can be used to apply template matching to the original image and template image. There are several comparison methods that can be used to create the resulting image. Please see the below link to look at the math behind each of these methods. The result is a grayscale image where each pixel denotes the similarity to the template image.

Template matching methods: https://docs.opencv.org/master/df/dfb/group__imgproc__object.html#ga3a7850640f1fe1f58fe91a2d7583695d

```python
result = cv2.matchTemplate(quarters, quarter_template, cv2.TM_CCOEFF)
plt.imshow(result, cmap='gray')
```

From this result, it becomes apparent that the greatest match occurs at around row 80, column 120. This is the location of the brightest pixel and will serve as the top left of the rectangle that will be drawn. Now the minMaxLoc() method can be used to find the location of this top left corner.
